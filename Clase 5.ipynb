```python
df = pd.read_excel('datasets/Canada.xlsx',
                   sheet_name = 'Canada by Citizenship', 
                   engine = 'openpyxl', 
                   skiprows = range(20), 
                   skipfooter = 2, 
                  )
df
```

```python
df.columns = list(map(str, df.columns))
all(isinstance(x, str) for x in df.columns)
```

```python
df.drop(['Type',    'Coverage','REG','DEV',
           'DevName','Unnamed: 43',
       'Unnamed: 44', 'Unnamed: 45', 'Unnamed: 46', 'Unnamed: 47',
       'Unnamed: 48', 'Unnamed: 49', 'Unnamed: 50'], inplace = True, 
       axis = 1 )
df.columns
```

```python
df.rename(columns = {'OdName': 'Country',
                     'AreaName': 'Continent',
                     'RegName': 'Region',},
          inplace = True)
df.columns
```

```python
df['Total']=df.iloc[:,4:].sum(axis=1)
df.head()
df.set_index('Country', inplace=True)
df.head()
```

# Me gustaria saber cuantas personas llegaron desde Haiti a Canada desde 1980 a 2013 con una grafica de lineas

```python
df.loc['Haiti', years].plot()
plt.xlabel('Years', fontsize = 16)
plt.xticks(fontsize = 16, rotation = 45)

plt.ylabel('Numero de imigrantes', fontsize = 16)
plt.yticks(fontsize = 16, rotation = 45)

plt.title('Numero de imigrantes desde Haiti a Canada desde 1980-2013',
          fontsize = 13)

plt.text(30,6500,'Terremoto')

plt.show()
```

```python
dfN = df.loc[['China','India'],years]
dfN = dfN.transpose()
dfN.head(3)
```

```python
dfN.plot(kind='line')
```

```python
df_topfive = df.sort_values(by= 'Total', 
                            ascending = False, 
                            axis = 0)
df_topfive = df_topfive.drop(['Total', 'Unknown'], axis = 0)
df_topfive = df_topfive.head(7)
df_topfive = df_topfive.loc[:, years]
df_topfive = df_topfive.transpose()

df_topfive.plot(kind = 'area', figsize = (20,10), alpha =0.5)
```
