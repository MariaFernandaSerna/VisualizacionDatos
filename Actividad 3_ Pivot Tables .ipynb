# Examen A3: Pivot Tables

```python
import pandas as pd 
import numpy as np
```

```python
%ls datasets
```

```python
df = pd.read_csv('datasets/antropometria-dataset.csv')
df.head()
```

```python
my_pivot = pd.pivot_table(df,
                          index = 'desc_ent',
                          aggfunc = 'mean',
                          
                          values = 'peso',
                         )
my_pivot
```

```python
my_pivot = pd.pivot_table(df,
                          index = 'desc_ent',
                          aggfunc = 'median',
                          columns = 'sexo',
                          values = 'peso',
                         )
my_pivot
```

```python
my_pivot = pd.pivot_table(df,
                          index = 'desc_ent',
                          aggfunc = 'mean',
                          columns = 'sexo',
                          values = 'peso',
                         )
my_pivot
```
